# 01_EDA: Eksploracyjna Analiza Danych

W tym notebooku przeanalizujemy dane z konkursu Digit Recognizer.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych
train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')
print(f'Train shape: {train.shape}')
print(f'Test shape: {test.shape}')
train.tail()
```

## Rozkład etykiet (label) w zbiorze treningowym

```python
sns.countplot(x='label', data=train)
plt.title('Rozkład etykiet w zbiorze treningowym')
plt.show()
```

## Przykładowe obrazy cyfr

```python
def plot_digits(data, labels, n=10):
    idxs = np.random.choice(len(data), n, replace=False)
    plt.figure(figsize=(15, 2))
    for i, idx in enumerate(idxs):
        plt.subplot(1, n, i+1)
        plt.imshow(data.iloc[idx, 1:].values.reshape(28, 28), cmap='gray')
        plt.title(labels.iloc[idx])
        plt.axis('off')
    plt.show()

plot_digits(train, train['label'], n=10)
```
